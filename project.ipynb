```python
df = pd.read_csv('/content/rcsb_pdb_custom_report_20240411062134.csv')
```

```python
df_cleaned = df.dropna(subset=['Entity ID'])
```

```python
df_cleaned = df_cleaned.drop(columns=['Unnamed: 3'])
```

```python
df_cleaned['Entity ID'] = df_cleaned['Entity ID'].str.split('_').str[0] + ':' + df_cleaned['Auth Asym ID']
```

```python
df_cleaned = df_cleaned.drop(columns=['Auth Asym ID'])
```

```python
df_cleaned
```

```python
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
len(df_cleaned)
```

```python
output = '\n'.join(df_cleaned['Entity ID'].values)
```

```python
file_name = "out.txt"

# Using the 'with' statement to ensure the file is properly closed after writing
with open(file_name, 'w') as file:
    file.write(output)

print(f"String saved to {file_name}")
```

```python
with open('/content/fasta.txt') as f:
    fastas = f.read().split('\n\n')
```

```python
fastas[:1]
```

```python
clean_list = []
for fasta in fastas:
    id = fasta.split()[0]
    sequence = ''.join(fasta.split('\n')[1:])
    clean_list.append((id, sequence))
```

```python
clean_list
```

```python
file_name = 'clean_fasta.txt'
```

```python
with open(file_name, 'w') as file:
    for item in clean_list:
        # Writing the sequence and the new line character to the file
        file.write(f"{item[0]}\n{item[1]}\n")
```
